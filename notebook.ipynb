### This cell is used to push data from mongo to Kafka

```python
import json
from pymongo import MongoClient
from kafka import KafkaProducer
from bson import ObjectId
import json

def error_callback(exc):
    raise Exception('Error while sendig data to kafka: {0}'.format(str(exc)))

def write_to_kafka(topic_name):
  client = MongoClient("mongodb://localhost:27017")
  db = client['input_data']
  collection = db["hihihaha"]
  producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'])
  for doc in collection.find():
    doc['_id']=str(doc['_id'])
    train_json = json.dumps(doc)
    producer.send(topic_name,  value=train_json.encode('utf-8')).add_errback(error_callback)
  producer.flush()
  producer.close()

```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col,length
from pyspark.sql.types import StringType
import json
from pyspark.sql import functions as F
from pyspark.sql.functions import col, from_json
from pyspark.sql.types import StringType, StructType
from kafka import KafkaProducer, KafkaConsumer
from concurrent.futures import ThreadPoolExecutor
import plotly.graph_objects as go
import plotly.io as pio
import matplotlib.pyplot as plt
from IPython.display import display
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objects as go
from pyngrok import ngrok
from jupyter_dash import JupyterDash 
import dash_cytoscape as cyto 
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Output, Input
import pandas as pd  
import plotly.express as px
import math
from dash import no_update
from pyngrok import ngrok

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = JupyterDash(__name__, external_stylesheets=external_stylesheets)
app.run_server(mode='external')

app.layout = html.Div(
    html.Div([
        html.H4('Sentiment Analysis Dashboard'),
        html.Div(id='live-update-text'),
        dcc.Graph(id='sentiment-pie-chart'),
        dcc.Graph(id='sentiment-bar-chart'),
        dcc.Interval(
            id='interval-component',
            interval=5*1000,  
            n_intervals=0
        )
    ])
)

@app.callback(
    Output('sentiment-pie-chart', 'figure'),
    Input('interval-component', 'n_intervals')
)
def callback1(n):
  return fig

@app.callback(
    Output('sentiment-bar-chart', 'figure'),
    Input('interval-component', 'n_intervals')
)
def callback2(n):
  return fig1
sia = SentimentIntensityAnalyzer()
def analyze_sentiment(tweet):
    sentiment_scores = sia.polarity_scores(tweet)
    compound_score = sentiment_scores['compound']
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
analyze_sentiment_udf = udf(analyze_sentiment, StringType())
def visual():
    my_spark = SparkSession \
    .builder \
    .config("spark.executor.memory", "1g") \
    .appName("csc14112") \
    .config("spark.mongodb.read.connection.uri", "mongodb://localhost:27017/tweetss.sentiment") \
    .getOrCreate()

    movies_lang_df = my_spark.read \
      .format('mongodb') \
      .option("database", "tweetss") \
      .option("collection", "sentiment") \
      .load()
    movies_lang_df.show()
    global fig
    global fig1
    sentiment_counts = movies_lang_df.select("Sentiment").groupBy("Sentiment").count().toPandas()

    fig = go.Figure(data=go.Pie(labels=sentiment_counts["Sentiment"], values=sentiment_counts["count"]))
    fig.update_layout(title="Sentiment Analysis", showlegend=False)
    
    
    fig1 = go.Figure(data=go.Bar(x=sentiment_counts["Sentiment"], y=sentiment_counts["count"]))
    fig1.update_layout(title="Sentiment Counts", xaxis_title="Sentiment", yaxis_title="Count")
    display(fig)
    display(fig1)
    print("capnhat1")
def write_row(batch_df):
    client = MongoClient("mongodb://localhost:27017")
    db = client["tweetss"]
    collection_name = "sentiment"
    collection = db[collection_name]
    data_list = batch_df.toPandas().to_dict(orient="records")
    if data_list:
        collection.insert_many(data_list)
        visual()
    else:
        print("Data list is empty. Skipping insertion.")
    client.close()
def consume_from_kafka():

    spark = SparkSession.builder \
        .appName("StreamingExample") \
        .master("local[*]") \
        .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.1.1")\
        .config("spark.driver.memory", "2g") \
        .getOrCreate()

    kafka_topic_name = "reco-train"
    kafka_bootstrap_servers = 'localhost:9092'

    schema = StructType().add("Date", StringType()).add("Tweet", StringType())


    df = spark.readStream.format("kafka") \
        .option("kafka.bootstrap.servers", kafka_bootstrap_servers) \
        .option("subscribe", kafka_topic_name) \
        .option("value.deserializer", "org.apache.kafka.common.serialization.StringDeserializer") \
        .load() 
    
    df = df.withColumn("value", col("value").cast(StringType()))
    df = df.withColumn("value", from_json(col("value"), schema))
    df = df.select("value.Date", "value.Tweet")
    df_with_sentiment = df.withColumn("Sentiment", analyze_sentiment_udf(col("Tweet")))

    query = df_with_sentiment.writeStream \
    .foreachBatch(lambda batch_df, _: write_row(batch_df)) \
    .trigger(processingTime='5 seconds') \
    .start()
    
    query.awaitTermination(timeout=50)


with ThreadPoolExecutor() as executor:
    executor.submit(write_to_kafka, 'reco-train')
    executor.submit(consume_from_kafka())
### RUN THIS CELL AGAIN IF IT DOES NOT WORK ###
### LINK DASHBOARD ON TOP CONSOLE OF THIS CELL ###
```
